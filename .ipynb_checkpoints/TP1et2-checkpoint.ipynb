```python
types = [ m['type'] for m in feed['data'] ]

def getTypeDistribution(types):
    tyred={}

    for t in types:
        if t in tyred:
            tyred[t] = tyred[t] + 1
        else:
            tyred[t] = 1

    tyred = { k: v/len(types) for k, v in tyred.items() }
    return tyred

tyred = getTypeDistribution(types)
x = np.arange(len(tyred))

fig, ax = plt.subplots(figsize=(12, 8))
plt.title("Post Type Distribution - 2018")
plt.xlabel("Post Type")
plt.ylabel("Post Frequency")
ax.bar(x, tyred.values())
ax.set_xticks(x)
ax.set_xticklabels(tyred.keys())

# plt.plot(range(1, max_depth), b_res, label='Bagging Regressor with DT')
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
